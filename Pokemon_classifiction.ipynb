```
import os
from keras.preprocessing import image
```

```
folders = ['Dratini', 'Aerodactyl' ,'Bulbasaur', 'Pikachu', 'Spearow', 'Psyduck', 'Meowth', 'Charmander', 'Squirtle']
```

```
labels_dict={}
for i in range(len(folders)):
  labels_dict[folders[i]]=i
```

```
labels_dict
```

```
x=[]
labels=[]
count=0
for pokemon in folders:
  p=os.listdir(pokemon)
  print(pokemon)
  for im in p:
    count+=1
    if(count>328):
      break
    name=os.path.join(pokemon,im)
    img = image.load_img(name,target_size=(224,224))
    img = image.img_to_array(img)
    x.append(img)
    labels.append(labels_dict[pokemon])
    
    
```

```
print(len(x))
```

```
import random
combined = list(zip(x,labels))
random.shuffle(combined)
x[:],labels[:]=zip(*combined)
```

```
from keras.utils import np_utils
```

```
labels = np_utils.to_categorical(labels)
```

```
import numpy as np
```

```
x = np.array(x)
labels = np.array(labels)
```

```
print(x.shape,labels.shape)
```

```
from keras.applications.resnet50 import ResNet50
```

```
from keras.layers import *
from keras.models import Model
import matplotlib.pyplot as plt
from keras.applications.mobilenet import preprocess_input,decode_predictions
```

```
model = ResNet50(input_shape=(224,224,3),include_top=False, weights='imagenet')
```

```
model
```

```
np.unique(labels)
print()
```

```
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256,activation='relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(3,activation='softmax')(d1)
model_new = Model(input=model.input,outputs=fc2)
model_new.summary()
```

```
from keras.optimizers import Adam
```

```
adam = Adam(learning_rate=0.003)
model_new.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```
for ix in range(len(model_new.layers)):
  print(ix,model_new.layers[ix])
```

```
for ix in range(173):
  model_new.layers[ix].trainable=False
```

```
hist = model_new.fit(x,labels,shuffle=True,epochs=10,batch_size=32,validation_split=0.2)
```
