```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df_x = pd.read_csv(io.BytesIO(uploaded['Logistic_X_Train.csv']))
df_y = pd.read_csv(io.BytesIO(uploaded["Logistic_Y_Train.csv"]))
```

```
x=df_x.values
y=df_y.values
```

```
plt.style.use("seaborn")
plt.scatter(x[:,0],x[:,1],c=y[:,0])
plt.show()
```

```
from keras import models
```

```
model = models.Sequential()
```

```
from keras.layers import Dense
```

```
model.add(Dense(16,activation='relu',input_shape=(x.shape[1],)))
```

```
model.add(Dense(16,activation='relu'))
```

```
model.add(Dense(1,activation='sigmoid'))
```

```
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```
model.summary()
```

```
x_val = x[:500]
x_train_new = x[500:]

y_val = y[:500]
y_train_new = y[500:]
```

```
hist = model.fit(x_train_new,y_train_new,epochs=35,batch_size=512,validation_data=(x_val,y_val))
```

```
h=hist.history
```

```
plt.plot(h['loss'],label='loss')
plt.plot(h['val_loss'],label='val_loss')
plt.legend()
plt.plot()
```

```
plt.plot(h['accuracy'],label='accuracy')
plt.plot(h['val_accuracy'],label='val_accuracy')
plt.legend()
plt.plot()
```
